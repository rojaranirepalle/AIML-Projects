<h4>Dictionary: keys are customer names and values are lists of ordered products</h4>

```python
# Use a dictionary where keys are customer names and values are lists of ordered products

customer_orders={}

for cust_name, product, price, category in orders:
    
    if cust_name not in customer_orders:
        customer_orders[cust_name] = [product]
    else:
        customer_orders[cust_name].append(product)

print(customer_orders)
```

<h2>2. Classify products by category</h2>

<h4>Map each product to its respective category</h4>

```python
# Use a dictionary to map each product to its respective category

product_category={}

for cust_name, product, price, category in orders:
    
    if product not in product_category:
        product_category[product] = category

print(product_category)
```

<h4>Unique product categories</h4>

```python
# Create a set of unique product categories

categories=set()

for cust_name, product, price, category in orders:
        categories.add(category)

print(categories)
```

<h4>Display all available product categories</h4>

```python
# Display all available product categories

print("All available product categories:")

for cat in categories:
    print(cat)
```

<h4>Display products under each category</h4>

```python
category_products={}

for cust_name, product, price, category in orders:
    
    if category not in category_products:
        category_products[category]= [product]
    else:
        if product not in category_products[category]:
            category_products[category].append(product)

print(category_products)
```

<h2> 3. Analyze customer orders </h2>

<h4>Total amount each customer spends</h4>

```python
# Use a loop to calculate the total amount each customer spends

customer_totalsales={}

for cust_name, product, price, category in orders:
    
    if cust_name not in customer_totalsales:
        customer_totalsales[cust_name]= price
    else:
        customer_totalsales[cust_name]+=price

print(customer_totalsales)
```

<h4>customer classification based on the total sale</h4>
<ul><li>If the total purchase value is above 100, classify the customer as a high-value buyer</li>
<li>If it is between 50 and 100, classify the customer as a moderate buyer</li>
<li>If it is below 50, classify them as a low-value buyer</li></ul>

```python
buyer_classification={}

for customer_name, total_sale in customer_totalsales.items():
    
    if total_sale > 100:
        category = "High-Value Buyer"
    elif 50 <= total_sale <= 100:
        category = "Moderate Buyer"
    else:
        category = "Low-Value Buyer" 
    buyer_classification[customer_name]= category
    
print(buyer_classification)
```

<h2>4. Generate business insights</h2>

<h4>Total revenue per product category</h4>

```python
#  Calculate the total revenue per product category and store it in a dictionary

category_totalsales={}

for cust_name, product, price, category in orders:
    
    if category not in category_totalsales:
        category_totalsales[category]= price
    else:
        category_totalsales[category]+=price

print(category_totalsales)
```

<h2>5. Organize and display data</h2>

<h4>Each customer’s total spending and their classification</h4>

```python
# Print a summary of each customer’s total spending and their classification
customer_totalsales = dict(sorted(customer_totalsales.items(), key=lambda item: item[1], reverse=True))

for customer in customer_totalsales:
    print(f"{customer} → Total: {customer_totalsales[customer]} → {buyer_classification[customer]}")
```

<h4>Customers who purchased from multiple categories</h4>

```python
# Use set operations to find customers who purchased from multiple categories
customers_mul_categories={}
categroies={}
for cust_name, product, price, category in orders:
        if cust_name not in categroies:
            categroies[cust_name]= [category]
        else:
            if category not in categroies[cust_name]:
                categroies[cust_name].append(category)
    
        if(len(categroies[cust_name])>=2):
            customers_mul_categories[cust_name]=categroies[cust_name]

if(len(customers_mul_categories)>=1):
    print("\nCustomers who purchased from multiple categories:")
    for cust_name in customers_mul_categories:
        print(cust_name)
```

<h4>Common customers who bought both Electronics and Clothing</h4>

```python
# Identify common customers who bought both Beverages and Grains

customers_electronics_categories=set()
customers_clothing_categories=set()

for cust_name, product, price, category in orders:
        if category == "Electronics":
            customers_electronics_categories.add(cust_name)
        if category == "Clothing":
            customers_clothing_categories.add(cust_name)

customers_both_categories=customers_electronics_categories.intersection(customers_clothing_categories)

if(len(customers_both_categories)>=1):
    print("Customers who bought both Electronics and Clothing:")
    for cust_name in customers_both_categories:
        print(f"{cust_name}")
```
