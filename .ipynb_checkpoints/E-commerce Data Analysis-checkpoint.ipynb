<center><h1> E-commerce Data Analysis </h1></center>

```python
#data taken from https://www.kaggle.com/datasets/ayushparwal2026/online-ecommerce?select=Online-eCommerce.csv

import csv
from tabulate import tabulate

orders=[]
with open("eCommerceData.csv","r") as ecom_file:
    reader=csv.DictReader(ecom_file, skipinitialspace=True)
    for row in reader:
        if any(row): 
            orders.append(row)
        
#Product Popularity Analysis
product_count = {}
category_count = {}
for order in orders:
    product = order["Product"]
    category = order["Category"]
    
    if product in product_count:
        product_count[product] += 1
    else:
        product_count[product] = 1
        
    if category in category_count:
        category_count[category] += 1
    elif(category!=''):
        category_count[category] = 1

product_count = dict(sorted(product_count.items(), key=lambda item: item[1], reverse=True))
category_count = list(sorted(category_count.items(), key=lambda item: item[1], reverse=True))
```

<h2> Sale wise Top 10 Products & Number of saled quanties </h2>

```python
#top 10 products
products_10=list(product_count.items())[:11]
print("Product Popularity:")
print(tabulate(products_10, headers=("Product", "Count"), tablefmt='grid'))
```

# Category wise number of product sales

```python
#top categories
print("Category Popularity:")
print(tabulate(category_count, headers=("Category", "Count"), tablefmt='grid'))
```
